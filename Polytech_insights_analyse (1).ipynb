```python
# définir une fonction pour recoder la variable 'AGE':
def age_groupes(x):
    if x >= 18 and x < 25:
        return "18-24"
    if x >= 25 and x < 35:
        return "25-34"
    if x >= 35 and x < 50:
        return "35-49"
    if x >= 50 and x < 65:
        return "50-64"
    if x >= 65 and x < 75:
        return "65-74"
    if x == 75 or x > 75:
        return "75 et plus"
```

```python
data['Age_groupes']= data['AGE'].apply(lambda x : age_groupes(x))
```

```python
# définir une fonction pour recoder la variable 'DIPL2':
def diplome(x):
    if x == "Vous n’avez pas été scolarisé(e)":
        return "Aucun diplôme, CEP"
    if x == "Aucun diplôme mais vous avez suivi une scolarité jusqu’en école primaire ou au collège":
        return "Aucun diplôme, CEP"
    if x == "Aucun diplôme mais vous avez suivi une scolarité au-delà du collège":
        return "Aucun diplôme, CEP"
    if x == "CEP (certificat d’études primaires)":
        return "Aucun diplôme, CEP"    
    if x == "BEPC, brevet élémentaire, brevet des collèges" or x == "CAP, brevet de compagnon" or x == "BEP":
        return "Avant Bac"
    if x == "Baccalauréat technologique ou professionnel, brevet professionnel ou de technicien, BEA, BEC, BEI, BEH, capacité en dr":
        return "Bac techno pro"
    if x == "Baccalauréat général, brevet supérieur":
        return "Bac"
    if x == "Diplôme de 1 er cycle universitaire , licence,BTS, DUT, diplôme des professions sociales ou de la santé, d'infirmier(":
        return "Diplôme de 1 er cycle universitaire , licence,BTS, DUT, diplôme des professions sociales ou de la santé, d'infirmier"
    if x == "Diplôme de 2 e ou 3 e cycle universitaire (y compris médecine, pharmacie, dentaire), master, diplôme d’ingénieur,":
        return "Diplôme de 2 e ou 3 e cycle universitaire (y compris médecine, pharmacie, dentaire), master, diplôme d’ingénieur"
```

```python
data['Diplome']= data['DIPL2'].apply(lambda x : diplome(x))
```

```python
# fonction pour creer un tableau croisé multiple avec effectifs et pourcentages
def tableau_croise_final(
    data, var_dep, indeps):
    t_nombre = pyshs.tableau_croise_multiple(data,
                              var_dep,
                              indeps,
                              "POIDS",               
                              contenu="absolu") 
    t_pourcentage = pyshs.tableau_croise_multiple(data,
                              var_dep,
                              indeps,
                              "POIDS",                   
                              contenu="pourcentage" )
    t=pd.concat({n:pd.concat({"Effectif":t_nombre[n],"%":t_pourcentage[n]},axis=1)                     
        for n in t_nombre.columns if n != "Total" and n != "Distribution"},axis=1)
    
    for n in t_nombre.columns:
        if n == "Total":
            t["Total"] = t_nombre["Total"]

    t["Distribution"]= t_pourcentage["Distribution"]
    
    t = t.reset_index()
    
    t["Question"] = t["Variable"].apply(lambda x : x.split(' ')[0])
    
    t["Question"] = t["Question"].apply(lambda x : list({key: questions_variables[key] for key in [x]}.values())[0])
    
    t = t.set_index(['Question','Variable'])
    
    return t


   
```
